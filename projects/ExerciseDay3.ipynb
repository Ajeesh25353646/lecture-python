```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
dataframe = pd.read_csv('lecture-python/data/WaveData.csv')
dataframe.head(10)
```

```python
# Indication: recuperer le nom des colonnes automatiquement
col_list = dataframe.columns.tolist()
print(col_list)
```

1. Faire la distribution (histogramme) de toutes les variables **sauf la date** avec matplotlib

```python
# Creer une nouvelle figure avant de faire l'histo
plt.figure()
plt.hist(np.random.normal(size=1000));
```

```python
# Creer une seule figure avec plusieurs sous-figures
def plot_df(df):
    plt.figure(figsize=(15, 7))
    for i, col in enumerate(col_list[1:]):
        plt.subplot(2, 3, i+1)
        plt.hist(df[col], bins=100)
        plt.xlabel(col)
```

2. Nettoyer les données et refaire la distribution de toutes les variables

```python
for col in col_list[1:]:
    dataframe = dataframe[dataframe[col]>-99]
```

```python
plot_df(dataframe)
```

```python
df1 = dataframe[dataframe['Peak Direction']<50]
df2 = dataframe[dataframe['Peak Direction']>=50]

plot_df(df1)
```

```python
plot_df(df2)
```

```python
def plot_dfs(list_df, list_labels):
    plt.figure(figsize=(15, 7))
    for df, label in zip(list_df, list_labels):
        for i, col in enumerate(col_list[1:]):
            plt.subplot(2, 3, i+1)
            plt.hist(df[col], bins=50, alpha=0.3, density=True, label=label)
            plt.xlabel(col)
            plt.legend()
    return
```

```python
plot_dfs([df1, df2], ['$<50$', '$\geq 50$'])
```
