# f-strings and formatting

```python
a =  "hello my name is' Louie and I am' 35 years old"
```

```python
print(a)
```

```python
age =35.567656755
```

```python
text =   "hello my name is' Louie and I am'"+ str(age)+" years old"

print(text)
```

```python
ftext = f"hello my name is Louie and I am {age} years old"
print(ftext)
```

```python
old_text = "hello my name is Louie and I am {} years old".format(age)
print(old_text)
```

```python
ftext = f"hello my name is Louie and I am {age*0.75} years old"
print(ftext)
```

```python
ftext = f"hello my name is Louie and I am {age*0.75:.3f} years old"
print(ftext)
```

```python
ftext = f"hello my name is Louie and I am {age*0.75:.3e} years old"
print(ftext)
```

```python
ages = { 'louie' : 35.67, 'paul' : 22.345, 'mathusela': 10000.4554}
for name, age in ages.items():
    print (f'The person called {name} is {age:.1f} years old')
```

```python
ages = { 'louie' : 35.67, 'paul' : 22.345, 'mathusela': 10000.4554}
for name in ages.keys():
    print (f'The person called {name} is {ages[name]:.1f} years old')
```

```python

ages = { 'louie' : 35.67, 'paul' : 22.345, 'mathusela': 10000.4554}
for name, age in ages.items():
    print ('The person called', name, ' is ', age , 'years old')
```
