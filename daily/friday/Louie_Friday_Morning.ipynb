```python
R = im[:, :, 0]
G = im[:, :, 1]
B = im[:, :, 2]
```

```python
binning = np.linspace(0,255, 254)
plt.hist(R_flat, bins=binning, color='red', alpha=0.1);
plt.hist(G_flat, bins=binning, color='green', alpha=0.1);
plt.hist(B_flat, bins=binning, color='blue', alpha=0.1);
```

```python
plt.hist([R_flat, G_flat, B_flat] , bins=binning, color= ['red', 'green', 'blue' ], stacked=True);
```

```python
im_zoomed =  im[250:1500,1500:2500,:]
```

```python
plt.imshow(im_zoomed)
```

```python
plt.hist([im_zoomed[:,:,i].flatten() for i in range(3) ] , bins=binning, color= ['red', 'green', 'blue' ], stacked=True);
```

```python
plt.imshow(im_zoomed-50);
```

```python
def safeAdd(array, val):
    result = array + val
    result[result < 0] = 0
    result[result > 255] = 200
    return result

    
```

```python
plt.imshow(safeAdd(im_zoomed,-50));
```

```python
plt.imshow(safeAdd(im_zoomed,50));
```

```python
def subtractArrays(array1, array2):
    result = array1 - array2
    result[result < 0] = 0
    result[result > 255] = 255
    return result
```

```python
plt.imshow(subtractArrays(im_zoomed, im_zoomed2))
```
