```python
# Usual import
import numpy as np
import matplotlib.pyplot as plt

# Load a test image
im = plt.imread('../../data/image_test.jpg')
```

```python
def plot_image(im, h=5, **kwargs):
    lx, ly = im.shape[:2]
    w = (ly/lx)*h
    plt.figure(figsize=(w,h))
    plt.imshow(im, interpolation=None, **kwargs)
    return
```

```python
plot_image(im)
```

### Questions

1. Create a new image in which all all *bright pixels* are turned to black
2. Create N random rectangle filled with random color in the picture

```python
def Bright2Black(ima, th=250):
    
    # Create a copy
    res = ima.copy()
    
    # Define the mask to select bright pixels
    R, G, B = res[..., 0], res[..., 1], res[..., 2]
    isBright = (R>th) & (G>th) & (B>th)
    
    # Replace the bright pixel with black
    res[isBright] = 0
    
    # Return the result
    return res
```

```python
im_2black = Bright2Black(im, th=220)
plot_image(im_2black)
```

```python
def with_colored_rectangle(ima, N=5):
    
    # Create a copy
    res = ima.copy()
    
    # Loop over rectangles
    for i in range(N):
    
        # Define random rectangles
        x  = np.random.randint(low=0, high=res.shape[0])
        dx = np.random.randint(low=0, high=res.shape[0]-x)
        y  = np.random.randint(low=0, high=res.shape[1])
        dy = np.random.randint(low=0, high=res.shape[1]-y)
    
        # Define random color
        color = np.random.randint(low=5, high=255, size=(3,), dtype=np.uint8)
    
        # Add the rectangles
        res[x:x+dx, y:y+dy] = color
    
    # Return the result
    return res
```

```python
plot_image(with_colored_rectangle(im, 5))
```
