# Reading in a file to numpy arrays

first objectives:
- read in the file "lecture-python-master/data/WaveData.csv"
    - Q1: How many lines are in this dataset?
- fill the data into a dictionary where the keys are the column labels and the values are numpy arrays of floats (or string for the date/time)
    - Q2:for each label, how many values are currupted ? (=-99)?
- make a mask which filters out any rows where any column is currupted
- calculate the mean and standard deviation of each quantity

```python
def read_file(filename):
    '''
    function which takes a filename and returns the contents as a list of lines
    '''
    file  = open(filename, 'r')
    result = []
    for line in file:
        result.append(line.strip())
    return result
    
```

```python
data = read_file('../data/WaveData.csv')
```

```python
print(f"the file had {len(data)} lines")
```

```python
data[0:10]
```

```python
headers=data[0]
```

```python
headers.split(',')
```

```python
data_dict = {}
```

```python
for header in headers.split(','):
    data_dict[header] = []
```

```python
data_dict
```

```python
keys = list(data_dict.keys())
keys
```

```python
float('1.234')
```

```python
for d in data[1:]:
    tokens = d.split(',')
    for i in range(len(keys)):
        if i == 0:
            data_dict[keys[i]].append(str(tokens[i]))
        else:
            data_dict[keys[i]].append(float(tokens[i]))
        #print(keys[i],data_dict[keys[i]] )
    
data_dict
```

```python
import numpy as np
for key in keys:
    #data_dict[key] =np.array(data_dict[key])
    data_as_list = data_dict[key]
    data_as_array = np.array(data_as_list)
    data_dict[key] = data_as_array 
```

```python
data_dict
```

```python
for key, value in data_dict.items():
    if key == 'Date/Time' : continue
    print(f"the column {key} has {sum(value==-99.9)} corrupted values")
```

```python
corruptedColumnsPerRow = np.zeros(len(data[1:]))
```

```python
corruptedColumnsPerRow
```

```python
for key, value in data_dict.items():
    if key == 'Date/Time' : continue
    corruptedColumnsPerRow += value==-99.9
```

```python
corruptedColumnsPerRow
```

```python
notCorruptedMask = corruptedColumnsPerRow == 0
```

```python
cleaned_dict = { key:value[notCorruptedMask] for key,value in data_dict.items()}
```

```python
len(cleaned_dict['Hs'])
```

```python
for key, value in cleaned_dict.items():
    if key == "Date/Time" : continue
    print(f'Feature "{key}" has mean={np.mean(value):.2f} +/- {np.std(value):.2f})')
```
