```python
# Input file and containers to store data
f = open('../../data/WaveData.csv', 'r')
features = []
waves = []

# Loop over the lines
for l in f:
    
    # Initialize the container for one wave
    w = {}
    
    # Clean the line and get each field
    values = l.strip().split(',')
    
    # Get the first line as being all the features
    if values[0] == 'Date/Time':
        features = values
    
    # Get the other line as data, as float except for the data
    else:
        for i, feat in enumerate(features):
            w[feat] = values[i]
        if feat != 'Date/Time':
                w[feat] = float(w[feat])
        waves.append(w)
```

```python
waves
```

```python
Data = {}
for feat in features[1:]:
    Data[feat] = []
    for w in waves:
        Data[feat].append(float(w[feat]))
```

```python
for k, v in Data.items():
    print(k, v[:5])
```

### Questions
 + how many corrupted waves
 + compute and print the mean and RMS of each features
 
### Correction

```python
# Total number of waves
Ntot = len(Data[features[1]])

# Loop over waves
Ncorr = 0
for iw in range(Ntot):
    for feat in features[1:]:
        if Data[feat][iw] < -99:
            Ncorr += 1
            break

print(Ncorr, Ntot, Ncorr/Ntot*100)
```

```python
# Define containers to store the result
# k:v -> k: feature, v: rms or mean
rms, mean = {}, {}

def get_mean(Xs):
    res, ntot = 0, len(Xs)
    for x in Xs:
        res += x/ntot
    return res
    
def get_rms(Xs):
    Xmean = get_mean(Xs)
    dX2 = [(x-Xmean)**2 for x in Xs]
    return get_mean(dX2)**0.5
    

# Loop over features
for feat, values in Data.items():
    rms[feat] = get_rms(values)
    mean[feat] = get_mean(values)
```

```python
for feat in Data.keys():
    print(f'{feat} : {mean[feat]:.2f} +/- {rms[feat]:.2f}')
```
