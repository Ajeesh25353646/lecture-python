- write your own function for to return `f(x) =  x*cos(a*x)`, with `a` as a parameter given as argument to the function.
- plot this function for 100 points between -6 and 6, and for 6 values of `a` between 1 and 3.

```python
def myFunc(x, a):
    return x*np.cos(x*a)
```

```python
x=np.linspace(-6,6,100)
```

```python
a_values = np.linspace(1 , 3, 6)
```

```python
for a in a_values:
    plt.plot(x, myFunc(x, a), label=f'a={a}')
    plt.legend()
```

## challenge

- write a function which calculates the discrete derivative F of a set of y-values
    
    $$ F = \frac{\Delta y }{ \Delta x} $$

 where $\Delta y_{i} = y_{i+1} - y_{i}$ and $\Delta x_{i} = x_{i+1} - x_{i}$ 

- you will notice tht this F has one point less in the array than thre length of y or x. you need to plot that on an array which represents the mid-points of x 

- plot the derivative of `g=x*cos(a*x)` for `a=2` along with `g`

```python
def delta(arr):
    #result = []
    #for i in range(len(arr)-1): result.append(arr[i+1] -arr[i])
    #return result
    return arr[1:] - arr[:-1]
```

```python
np.array(range(10))
```

```python
delta(np.array(range(10)))
```

```python
def deriv(x, y):
    return delta(y)/delta(x)
```

```python
len(myFunc(x,2))
```

```python
len(deriv(x, myFunc(x,2)))
```

```python
def midx(arr):
    return (arr[1:] + arr[:-1])/2
```

```python
plt.plot(x, myFunc(x, 2), label='g(x)')
plt.plot(midx(x), deriv(x, myFunc(x, 2)), label='g\'(x)')
plt.legend()
```
