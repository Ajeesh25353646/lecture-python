#### Step 1

Display the sin, cos and tan of the picture

```python
# Load the picture
im = plt.imread('Downloads/image_test.jpg')
plt.imshow(im)
```

```python
print(im[0, 0, :])
b = (np.sin(im[0, 0, :])+1)/2
print(b, b.dtype)
plt.imshow(b[np.newaxis, np.newaxis, :].astype(np.float32))
```

```python
def imsin(x):
    return (np.sin(x)+1)/2

Lambda = 255/(2*np.pi)
im_cos = (np.cos(im/Lambda)+1)/2
im_sin = (np.sin(im/Lambda)+1)/2

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(im_cos.astype(float))
plt.subplot(1, 2, 2)
plt.imshow(im_sin.astype(float));
```

```python
x = np.linspace(0, 255, 1000)
plt.plot(x, imsin(x/Lambda)*255);
```

```python
plt.plot(im[500, 1250:2250, 0], 'o');
```

```python
plt.plot(imsin(im[500, 1250:2250, 0]), 'o');
```

```python
plt.figure(figsize=(12, 4))
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.plot(imsin(im[1000, :, i]), 'o');
```

#### Step 2
Add sone random noise to the cathedrale  
Hint : You can generate random noise using np.random functions

```python
im_mod = im.copy()

im_cat = im_mod[500:1500, 1500:2500, :]

plt.figure()
plt.imshow(im_cat);
```

```python
im_cat_noise = im_cat + np.random.randint(-50, 50, size=im_cat.shape)

plt.figure()
plt.imshow(im_cat_noise.astype(np.uint8));
```

```python
im_mod[500:1500, 1500:2500, :] = im_cat_noise

plt.figure()
plt.imshow(im_mod);
```

## Step 3
Duplicate the cathedrale (as many time as you want)

```python
im_mod = im.copy()

im_cat = im_mod[500:1500, 1500:2500, :]

trans = [1000,1000]

im_mod[500+trans[0]:1500+trans[0], 1500+trans[1]:2500+trans[1], :] = im_cat

plt.figure()
plt.imshow(im_mod);
```
