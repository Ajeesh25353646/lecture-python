1. Mettre un carr√© de l'image en noir et blanc uniquement

```python
def get_channels(im):
    return im[..., 0], im[..., 1], im[..., 2]
```

```python
r, v, b = get_channels(image)
```

```python
def create_bw(r, v, b):
    return 0.3*r + 0.6*v + 0.1*b

bw = create_bw(r, v, b)
```

```python
plt.imshow(bw, cmap='Greys_r');
```

```python
im = image.copy()
imin, imax, jmin, jmax = 500, 1500, 1200, 2500
#for ic in range(3):
#    im[imin:imax, jmin:jmax, ic] = bw[imin:imax, jmin:jmax]
```

```python
im[imin:imax, jmin:jmax] = np.stack([bw[imin:imax, jmin:jmax]]*3, axis=2)
```

```python
%matplotlib inline
plt.figure(figsize=(10, 10))
plt.imshow(im);
```

```python
def is_in_circle(im, center, R):
    x0, y0 = center
    lx, ly = im.shape[:2]
    X, Y = np.meshgrid(np.arange(0, ly), np.arange(0, lx))
    radius = ((X-x0)**2 + (Y-y0)**2)**0.5
    return radius<=R
```

```python
im_test = np.random.randint(low=0, high=255, size=(10, 10, 3))
isIn = is_in_circle(im_test, [5, 5], 500)

plt.imshow(isIn)
```

```python
im = image.copy()
isIn = is_in_circle(im, [1900, 1100], 500)
im[isIn] = np.stack([bw[isIn]]*3, axis=1)
```

```python
print('im  : {}'.format(im.shape))
print('isIn: {}'.format(isIn.shape))
print('im[isIn]: {}'.format(im[isIn].shape))
```
