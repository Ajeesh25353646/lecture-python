```python
from scipy import optimize
from scipy import stats
```

```python
Npoints, Nsampling = 20, 1000

xcont = np.linspace(-5.0, 3.5, Nsampling)
x = np.linspace(-5, 3.0, Npoints)
y = 2*(np.sin(x/2)**2 + np.random.random(Npoints)*0.3)
dy = np.sqrt(0.10**2 + (0.10*y)**2)
```

```python
print(x[:5])
print(y[:5])
```

```python
plt.errorbar(x, y, yerr=dy, markersize=10, marker='o', linewidth=2, linestyle='');
```

```python
def model(s, p0, p1):
    return p0 + s*p1

best_parameters, _ = optimize.curve_fit(model, x, y, sigma=dy)
```

```python
print(best_parameters)
```

```python
ymodel = model(xcont, *best_parameters)
plt.errorbar(xcont, ymodel)
plt.errorbar(x, y, yerr=dy, markersize=10, marker='o', linewidth=2, linestyle='');
```

```python
ypredit = model(x, *best_parameters)
distance2 = np.sum(((ypredit-y)/dy)**2)
print(distance2)
```

```python
def model2(s, p0, p1, p2, p3, p4):
    return p0 + p1*s + p2*s**2 + p3*s**3 + p4*s**4
```

```python
best_parameters, _ = optimize.curve_fit(model2, x, y, sigma=dy)
print(best_parameters)
```

```python
ypredit = model2(x, *best_parameters)
distance2 = np.sum(((ypredit-y)/dy)**2)
print(distance2)
```

```python
ymodel = model2(xcont, *best_parameters)
plt.errorbar(xcont, ymodel)
plt.errorbar(x, y, yerr=dy, markersize=10, marker='o', linewidth=2, linestyle='');
```

```python
def fit_model(model):
    
    # Fit the function
    best_parameters, _ = optimize.curve_fit(model, x, y, sigma=dy)
    
    # Plot the result
    ymodel = model(xcont, *best_parameters)
    plt.errorbar(xcont, ymodel)
    plt.errorbar(x, y, yerr=dy, markersize=10, marker='o', linewidth=2, linestyle='')
    
    # Calcul du chi2
    ypredit = model(x, *best_parameters)
    distance2 = np.sum(((ypredit-y)/dy)**2)
    
    return distance2
```

```python
fit_model(model2)
```

```python
def model3(s, A, B, T, C):
    return A*np.sin(B*s+T) + C

fit_model(model3)
```
